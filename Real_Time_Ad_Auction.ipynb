```python
# STEP 2: Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
# STEP 3: Create dataset (reliable version of advertising dataset)
data = {
    'Daily_Time_Spent_on_Site': np.random.normal(65, 15, 1000),
    'Age': np.random.randint(18, 70, 1000),
    'Area_Income': np.random.normal(55000, 15000, 1000),
    'Daily_Internet_Usage': np.random.normal(180, 50, 1000),
    'Male': np.random.randint(0, 2, 1000)
}

df = pd.DataFrame(data)

df.head()
```

```python
#Normalize features
scaler = MinMaxScaler()

df[['time_norm', 'usage_norm']] = scaler.fit_transform(
    df[['Daily_Time_Spent_on_Site', 'Daily_Internet_Usage']]
)
```

```python
#Simulate CTR
df['ctr_prob'] = (
    0.5 * df['time_norm'] +
    0.3 * df['usage_norm'] +
    0.2 * np.random.rand(len(df))
)
```

```python
#Bidding Stategy
df['bid'] = df['ctr_prob'] * 10
```

```python
#Auction Ranking
df['rank'] = df['bid'].rank(ascending=False)

top_ads = df.sort_values(by='rank').head(10)
top_ads[['ctr_prob', 'bid', 'rank']]
```

```python
#Budget Optimization
budget = 50

df_sorted = df.sort_values(by='bid', ascending=False)
df_sorted['cumulative_cost'] = df_sorted['bid'].cumsum()

selected_ads = df_sorted[df_sorted['cumulative_cost'] <= budget]

print("Selected Ads:", len(selected_ads))
```

```python
#Performance Comparison
print("Avg CTR (Selected Ads):", selected_ads['ctr_prob'].mean())
print("Avg CTR (All Ads):", df['ctr_prob'].mean())
```
